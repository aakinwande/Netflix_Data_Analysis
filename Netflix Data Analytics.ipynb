```python
#Show the Show_id and Directors of "House of Cards"
netflixdataset[netflixdataset['Title'] == 'House of Cards']
```

```python
#In which year was the highest number of TV Shows and Movies released?, Show in Bar Graph

#Convert the Release_Date to a datetime format
netflixdataset['Date'] = pd.to_datetime(netflixdataset['Release_Date'], errors='coerce')
netflixdataset
```

```python
#Count the year

yearcount = netflixdataset['Date'].dt.year.value_counts()
yearcount
```

```python
#Bar Graph
yearcount.plot(kind= 'bar')
```

```python
#How many Movies and TV Shows are in the Dataset, Show in a Bar Graph
unique_category = netflixdataset.groupby('Category').Category.value_counts()
unique_category
```

```python
colors = ['skyblue', 'orange']
unique_category.plot(kind='bar', color=colors)
```

```python
#Show all the Movies that were released in the year 2000

#Create a year column from the Date column

netflixdataset['Year'] = netflixdataset['Date'].dt.year
```

```python
#Movies Released in 2000
netflixdataset[ (netflixdataset['Category'] == 'Movie') & (netflixdataset['Year'] == 2000.0) ]
```

```python
#Show only the Titles of all TV Shows release in Nigeria only
netflixdataset[ (netflixdataset['Category'] == 'TV Show') & (netflixdataset['Country'] == 'Nigeria') ]
```

```python
#Show the top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix
top_directors = netflixdataset['Director'].value_counts().head(10)
top_directors
```

```python
#Show all the records where "Category is Movie and Type is Comedies" or "Country is United Kingdom"

netflixdataset[ ((netflixdataset['Category'] == 'Movie') & (netflixdataset['Type'] == 'Comedies')) | (netflixdataset['Country'] == 'United Kingdom')]
```

```python
#How many Movies or TV Show was 'Tom Cruise' cast in

#Drop all null values
new_netflixdataset = netflixdataset.dropna()
new_netflixdataset[new_netflixdataset['Cast'].str.contains('Tom Cruise')]
```

```python
#What are the different/Unique ratings defined by Netflix?
netflixdataset['Rating'].unique()
```

```python
#What are the number of different/Unique ratings defined by Netflix?
netflixdataset['Rating'].nunique()
```

```python
#How many Movies got 'TV-14' rating in Canada?
netflixdataset[(netflixdataset['Category'] == 'Movie') & (netflixdataset['Rating'] == 'TV-14') & (netflixdataset['Country'] == 'Canada')]
```

```python
#How many Movies got 'R' Rating after the year 2018?
netflixdataset[(netflixdataset['Category'] == 'Movie') & (netflixdataset['Rating'] == 'R') & (netflixdataset['Year'] > 2018)]
```

```python
#Which Individual Country has the highest number of tv shows
tvshows_data = netflixdataset[netflixdataset['Category'] == 'TV Show']
tvshows_data.Country.value_counts().head(1)
```

```python
#Sort the dataset by Year
netflixdataset.sort_values(by = 'Year')
```
